# Array (массив)

Массив представляет полседовательный набор объектов.

image.png

 Отличительной особенностью является то, что он занимает конкретный участок в памяти, и заранее определено, сколько элементов в нем будет находиться. 
Это как и плюс так и минус одновременно. Плюс в том, что мы знаем позицию каждого элемента, и можем получить его фактически мгновенно.
Минусом является то, чтобы добавить конкертный элемент в массив, придется создавать новый массив на одну ячейку больше, перекидывать значения из старого массива, добавлять новый элемент и удалять старый массив.
