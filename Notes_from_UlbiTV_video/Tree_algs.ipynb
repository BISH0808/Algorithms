Задача: есть дерево, надо найти сумму значений, который хранит кажый узел

Реализуем функцию с помощью **рекурсивного подхода**

```python
const tree = [
    {
        v: 5,
        c: [
            {
                v: 10,
                c: [
                    {
                        v: 11
                    }
                ]
            },
            {
                v: 7,
                c: [
                    {
                        v: 5,
                        c: [
                            {
                                v: 1
                            }
                        ]
                    }
                ]   
            }
        ]
    }
];

const recursive = (tree) =>{
    let sum = 0;
    tree.forEach(node =>{
        sum += node.v
        if(!node.c){
            return node.v
        }
        sum += recursive(node.c)
    })
    return sum
}


console.log(recursive(tree))
```

Реализуем функцию с помощью **итеративного подхода**

Чтобы реализовать такую функцию, надо знать, что такое стек.
Стек очень похоже на очередь, но зедсь мы добавляем и извлекаем элеементы из конца. В очереди мы добавляли в конец, но извлекали из начала.

image.png

```python
const tree = [
    {
        v: 5,
        c: [
            {
                v: 10,
                c: [
                    {
                        v: 11
                    }
                ]
            },
            {
                v: 7,
                c: [
                    {
                        v: 5,
                        c: [
                            {
                                v: 1
                            }
                        ]
                    }
                ]   
            }
        ]
    },
    {
        v: 5,
        c: [
            {
                v: 10
            },
            {
                v: 15
            }
        ]
    },
];

const iteration = (tree) =>{
   if (!tree.length){
        return 0
   }
   let sum = 0
   let stack = []
   tree.forEach(node => stack.push(node));
   while (stack.length){
    const node=stack.pop()
    sum += node.v
    if (node.c){
       node.c.forEach(child=> stack.push(child)) 
    }
    
   }
   return sum
}

console.log(iteration(tree))
```
