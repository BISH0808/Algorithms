#Суть

Бинарное дерево - это структура данных, где каждый узел также является деревом, т.е. структура рекурсивна. Основная суть в том, что у каждого узла может быть только 2 потомка.
Добавляются эти узлы особым способом: если добавляемое в дерево значение меньше по значению, чем текущий узел, то значение уходит в левое поддерерво, если больше - в правое. Сравнение происходит с каждым узлом. и Так происходит, что правая часть выстраивается с большими значениями, а левая с меньшими.

Бинарное дерево поиска  аналогично бинарному алгоритму поиска.
