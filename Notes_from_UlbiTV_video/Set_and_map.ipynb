#Set and map

Стрктуры set и map хранят в себе структуры "ключ-значение"

Set хранит в себе только уникальные значения

image.png

#Код

Пример с **map**

```python
const map = new Map()
const objKey = {id:5}
map.set(objKey, "Ildar")

console.log(map.get(objKey));

```

**Ответ:** Ildar

Пример с **set**

```python
const map = new Map()
const objKey = {id:5}
map.set(objKey, "Ildar")

console.log(map.get(objKey));

const set = new Set()

set.add(5)
set.add(5)
set.add(5)
set.add(5)
set.add(4)
set.add(3)
set.add(2)

console.log(set)
```

**Ответ:**

Ildar
