Основная задача - понять существует ли путь из A в G.
Функия должна находить кратчайший путь, если таков имеется

# Структура данных "Очередь"

Очередь - это структура данных, состоящая из каких-то элементов. 
Основной принцип заключается в том, что элементы добавляются в конец структуры, а извлекаются всегда из ее начала.

Такой принцип называется **FIFO** (First In First Out).

image.png

#Код

```python
// Поиск в ширину в графе

const graph = {}
graph.a = ['b', 'c']
graph.b = ['f']
graph.c = ['d', 'e']
graph.d = ['f']
graph.e = ['f']
graph.f = ['g']

function breadthSearch(graph, start, end) {
    let queue = []
    queue.push(start)
    while (queue.length > 0) {
        // shift - это функция, которая извлекает первый элемент с массива
        const current = queue.shift()
        if (!graph[current]){
            graph[current] = []
        }
        if(graph[current].includes(end)){
            return true
        } else {
            queue = [...queue, ...graph[current]]
        }
    }

}

console.log(breadthSearch(graph, 'a', 'g'))
```

**Ответ:**

True
