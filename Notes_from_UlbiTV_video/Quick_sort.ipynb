# Суть алгоритма

Агоритм работает по так называемому принципу "разделяй и властвуй".
Суть в том, что мы делим массив на подмассивы и каждый раз рекурсивно мы выбираем опорный элемент у каждого массива. Его можно выбрать случайно, но чаще всего берут центральный.

Пробегаемся по массиву, и все элементы, которые меньше, чем опорный, добавляем в один массив, те, которые больше, в другой массив. 

Дальше для новых массивов выполняется такая же операция: выбирается опорный массив, происходит сортировка.

И так делается до тех пор, пока длина массива не станет равной 1.
Именно это условие будет базовым случаем выхода из рекурсии. 

Затем эти маленькие подмассивы склеиваются в один большой.
