#Суть

Суть свяазанного списка состоит в том, что каждый предыдущий элемент хранит ссылку на следующий элемент в списке.

Плюс такого подхода в том, что мы можем бесконечно добавлять элементы в конец или начало списка. А минус заключается в том, чтобы получить какой-то элемент нам надо с начала списка итерироваться и сравнивать.

Массив подходит в том случае, если мы часто обращаемся к данным.

Список же подходит в том случае, если мы рекдо обращаемся к данным, но часто его дополняем.
